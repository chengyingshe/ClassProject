### Split dataset

```python
def get_split(split_txt_file):
    with open(split_txt_file, 'r') as file:
        lines = file.readlines()
    train_ids = lines[0].strip()
    test_ids = lines[1].strip()
    return train_ids, test_ids

get_split('data/split_1.txt')
```
